```python
def get_delta(stats, setting_tuple, i, metric, n_samples, n_questions):
    setting1, setting2 = setting_tuple
    return setting1 + "_delta", [
            np.array(stats[setting1][str(i)][str(j)][metric][:n_samples]) \
            - np.array(stats[setting2][str(i)][str(j)][metric][:n_samples]) \
           for j in range(n_questions)
    ]
    

def get_deltas(stats, n_parts, metric, n_samples=None):
    """Compute deltas on example-by-example basis.
    Stats = dict where k= problem idx and v = stats for that problem"""
    deltas = {}

    setting_tuples = [
        ("oracle_comment", "baseline"),
        ("oracle_comment_w_full_q", "baseline_w_full_q"),
        ("self_comment", "baseline"),
        ("self_comment_w_full_q", "baseline_w_full_q"),
    ]


    n_questions = len(stats["oracle_comment"]["0"])
    tuple_names = []
    for i in range(n_parts):
        deltas[i] = {}

        for setting_tuple in setting_tuples:
            tuple_name, delta = get_delta(stats, setting_tuple, i, metric, n_samples, n_questions)
            deltas[i][tuple_name] = delta
            tuple_names.append(tuple_name)
        
    
    agg_deltas = {}
    for i in range(n_parts):
        agg_deltas[i] = {}
        for tuple_name in tuple_names:
            agg_deltas[i][tuple_name] = np.mean(deltas[i][tuple_name])

    return deltas, agg_deltas


deltas, agg_deltas = get_deltas(stats, 4, "test_case_average", n_samples=4)
```
